```python

from pyspark.sql import SparkSession
spark=SparkSession.builder.master("local[*]").getOrCreate()
spark=SparkSession.builder.appName("sol").getOrCreate()
```

```python
sc=spark.sparkContext
```

```python

#Exercice 1
import collections
lines = sc.textFile("/content/u.data")
ratings = lines.map(lambda x:x.split()[2])
result = ratings.countByValue()

sortedResults = collections.OrderedDict(sorted(result.items()))
for key, value in sortedResults.items():
  print("%s %i" % (key, value))
```

```python
#Exercice 2
def parseLine(line):
  fields=line.split(',')
  stationID=fields[0]
  entryType=fields[2]
  temperature=float(fields[3])*0.1*(9.0/5.0)+23.0
  return(stationID,entryType,temperature)
lines=sc.textFile("/1800.csv")
parsedLines=lines.map(parseLine)
minTemps=parsedLines.filter(lambda x:"TMIN"in x[1])
stationTemps=minTemps.map(lambda x:(x[0],x[2]))
minTemps=stationTemps.reduceByKey(lambda x,y:min(x,y))
results=minTemps.collect()
for result in results:
  print(result[0]+"\t{:.2f}F".format(result[1]))
```

```python
#Exercice 3
input=sc.textFile("/Book.txt")
words=input.flatMap(lambda x:x.split())
wordscounts= words.countByValue()
for word, count in wordscounts.items():
  cleanWord=word.encode('ascii','ignore')
  if(cleanWord):
    print(cleanWord.decode()+" "+ str(count))
```

```python

#Exercice 4
input=sc.textFile("/content/Book.txt")
words=input.flatMap(lambda x: x.split())
wordcounts=words.map(lambda x:(x,1)).reduceByKey(lambda x,y:x+y)
wordcountsSorted=wordcounts.map(lambda x:(x[1],x[0])).sortByKey()
results=wordcountsSorted.collect()
for  result in results:
  count=str(result[0])
  word=result[1].encode("ascii",'ignore')
  if(word):
    print(word.decode()+":\t\t"+count)
```

```python
# Exercice 8
# Définir la fonction de transformation pour parser chaque ligne du fichier
def parseline(line):
    fields = line.split(',')
    age = int(fields[2])
    num_friends = int(fields[3])
    return (age, num_friends)

# Charger le fichier CSV en tant que RDD
friends_lines = sc.textFile("/content/friends.csv")

# Appliquer la transformation pour obtenir un RDD de tuples (âge, nombre d'amis)
fields_rdd = friends_lines.map(parseline)

# Utiliser une transformation pour ajouter le nombre d'amis et le nombre d'occurrences par âge
age_friends_count = fields_rdd.mapValues(lambda x: (x, 1)).reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))

# Calculer la moyenne en divisant la somme du nombre d'amis par âge par le nombre d'occurrences
average_by_age = age_friends_count.mapValues(lambda x: x[0] / x[1])

# Collecter les résultats et les afficher
results = average_by_age.collect()
for result in results:
    print(f"Age: {result[0]}, Moyenne d'amis: {result[1]}")


```
